```python
import cfdadvection.advection as advec # Importing module  
import matplotlib.pyplot as plt
```

### Top hat function 

```python
# .Advection is the main class that has as __init__ function the:
# __init__(self, shape, n_cells, p_veloc_, num_periods)
# where shape is the shape of the initial profile eg: top_hat, gaussian
# n_cells is the number of cells 
# p_veloc is the perturbation velocity
# num_periods is the number of periods

first = advec.Advection("top_hat",256 , 1 , 5) 
                    #(shape, n_cells, p_veloc, num_periods)

a1 = first.forms()  #.form function gives the Exact solution

#the principal function is "advect" uses different functions inside the class
# It has the form advect(method, CFL)
# where method: slope limiter method used, CFL:Courant-Friedrichs-Lewy number 
# the different functions inside advect() are: 
# fill_bcs: fill the boundary conditions by using 
# time_step: get the time step
# states: compute interface states by using an specific slope limiter
# riemman: solve the Riemman problem at all interfaces
# convservative update: Does the conservative update 

#Advect top_hat profile using godunov approach with CFL = 0.5
a2 = first.advect("godunov", 0.5) 
#Advect top_hat profile using centered difference method with CFL = 0.5
a3 = first.advect("unlimited", 0.5) 
#Advect top_hat profile usind minmod slope limiter with CFL = 0.5
a4 = first.advect("minmod", 0.5) 
#Advect top_hat profile usind MC slope limiter with CFL = 0.5
a5 = first.advect("mc", 0.5) 

# The function .plot() helps us giving us the actual x and y values to plot
x1, y1 = first.plot(a1)
x2, y2 = first.plot(a2)
x3, y3 = first.plot(a3)
x4, y4 = first.plot(a4)
x5, y5 = first.plot(a5)

# Begining of the plot

plt.figure(figsize=(10,9))

plt.plot(x1,y1, color = "magenta", linestyle = "--", label = r"${\rm exact}: t=0$")
plt.plot(x2, y2, linestyle = "--", label = r"${\rm godunov}: t=5 T$")
plt.plot(x3, y3, linestyle = "--", label = r"${\rm unlimited}: t=5 T$")
plt.plot(x4, y4, linestyle = "--", label = r"${\rm minmod}: t=5 T$")
plt.plot(x5, y5, linestyle = "--", label = r"${\rm MC}: t=5 T$")


ax = plt.gca()
ax.tick_params(axis='both', which='major', labelsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
plt.xlabel('x', size=22)
plt.ylabel('a', size=22)

plt.legend(fontsize=13)
plt.show()
```
